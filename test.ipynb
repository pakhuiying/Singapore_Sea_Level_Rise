```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def f(x):
    return np.sin(np.sqrt(1*x ** 2))

n = 10

x = np.linspace(-1, 5, n)

plt.figure()
plt.plot(x,f(x))
plt.show()
```

```python
from matplotlib.patches import Rectangle
n = 10
X = np.ones((n,n))
Z = X.copy()
Z[np.s_[-3:,:]] = 0
Z[np.s_[:,-3:]] = 0

fig, ax = plt.subplots(1,1)
ax.imshow(Z)
ax.set_xlim(0,n)
ax.set_ylim(n,0)
ax.add_patch(Rectangle((0, 7), 7, 3, fill=True, hatch='+',color='green',alpha=0.5))
ax.add_patch(Rectangle((7, 0), 3, 10, fill=True, hatch='+',color='green',alpha=0.5))
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

from matplotlib import cbook, cm
from matplotlib.colors import LightSource

# Load and format data
dem = cbook.get_sample_data('jacksboro_fault_dem.npz')
z = dem['elevation']
nrows, ncols = z.shape
x = np.linspace(dem['xmin'], dem['xmax'], ncols)
y = np.linspace(dem['ymin'], dem['ymax'], nrows)
x, y = np.meshgrid(x, y)

region = np.s_[5:50, 5:50] # returns a tuple of slice objects, useful to indx array afterwards
x, y, z = x[region], y[region], z[region]
z = z - z.min()
# Set up plot
fig, ax = plt.subplots(subplot_kw=dict(projection='3d'))

ls = LightSource(270, 45)
# To use a custom hillshading mode, override the built-in shading and pass
# in the rgb colors of the shaded surface calculated from "shade".
rgb = ls.shade(z, cmap=cm.gist_earth, vert_exag=0.1, blend_mode='soft')
surf = ax.plot_surface(x, y, z, rstride=1, cstride=1, facecolors=rgb,
                       linewidth=0, antialiased=False, shade=False)

plt.show()
```
